### Now, let’s build the model. Let's import the linear regression class from scikit-learn.

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Assuming 'df' is your dataframe

# ... (your existing code to load and prepare 'df') ...

# Separate the target variable
y = df["Exam_Score"]
x = df.drop("Exam_Score", axis=1)

# Identify columns with object (string) dtype
categorical_cols = x.select_dtypes(include=['object']).columns

# Convert categorical features to numerical using one-hot encoding
# Assuming 'Parental_Involvement_in_Studies' is the actual categorical column you want to encode
# Replace 'Parental_Involvement_in_Studies' with the actual column name from categorical_cols
x = pd.get_dummies(x, columns=categorical_cols, drop_first=True)
# drop_first=True to avoid multicollinearity

# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.80, random_state=1)

# Create and fit the model
lr = LinearRegression()
lr.fit(x_train, y_train)

# ... (rest of your code) ...
````

# **Model Evaluation**

####Let’s take a look at the performance of the model. To do this, I’m going to use the coefficient of determination. The closer this value is to 1, the better the model. First, let’s take a look at the score of the model on the test data.


````python
lr.score(x_test,y_test)
````

#### ```The coefficient of determination (R²) is a number between 0 and 1 that measures how well a statistical model predicts an outcome.```

##### The coefficient of determination on the test data is greater than 0.7. Our model is not bad. Of course, it would be better if it was closer to 1

### Now, let’s see the score of the model on the training data.

````python
lr.score(x_train,y_train)
````

##### If the performance of the model on the training data was high, it would mean that there is an overfitting problem

####To overcome the overfitting problem, we can use regularization. Ridge or lasso models can be used for this.

### Now let’s take a look at another metric, mean squared error, to evaluate the model. For this, let’s first predict the test data with the predict method.



````python
y_pred = lr.predict(x_test)
````

````python
from sklearn.metrics import mean_squared_error
import math
math.sqrt(mean_squared_error(y_test, y_pred))
````
